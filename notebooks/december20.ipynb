```python
def reverse(s):
    return "".join(reversed(s))

class Tile:
    def __init__(self, tile_id, edges, values):
        self.id = tile_id
        self.edges = edges
        self.values = values
    
    def rotate(self, times):
        edges = [self.edges[(i + times) % 4] for i in range(4)]

        its = [(0, 1), (1, 0), (0, -1), (-1, 0)]
        n = len(self.values)
        origins = [(0, 0), (0, n - 1), (n - 1, n - 1), (n - 1, 0)]
        it0 = its[(times + 1) % 4]
        it1 = its[times % 4]
        origin = origins[times % 4]
        values = [[0] * n for i in range(n)]
        for i in range(n):
            pos = origin[0] + i * it0[0], origin[1] + i * it0[1]
            for j in range(n):
                values[i][j] = self.values[pos[0]][pos[1]]
                pos = pos[0] + it1[0], pos[1] + it1[1]

        return Tile(self.id, edges, values)

    def flip(self):
        edges = [reverse(self.edges[0]),
                 reverse(self.edges[3]),
                 reverse(self.edges[2]),
                 reverse(self.edges[1])]
        
        values = [list(reversed(row)) for row in self.values]
        
        return Tile(self.id, edges, values)

def parse_tile(strings):
    tile_id = int(strings[0][5:-1])
    edges = [strings[1],
             "".join(s[-1] for s in strings[1:]),
             reverse(strings[-1]),
             reverse("".join(s[0] for s in strings[1:]))]
    values = [[c for c in s[1:-1]] for s in strings[2:-1]]
    return Tile(tile_id, edges, values)
```

```python
def make_edge_dict(tiles):
    edges = {}
    for tile in tiles:
        for i, edge in enumerate(tile.edges):
            edges.setdefault(edge, []).append((tile.id, False))
            edges.setdefault(reverse(edge), []).append((tile.id, True))
    return edges
EDGES = make_edge_dict(ALL_TILES)
```

```python
def find_tiles(image, i, j, tile_dict, edge_dict, unavailable_tile_ids):
    edge_neighbors = None, None
    if i == 0:
        if 0 < j:
            edge_neighbors = (3, (i, j-1, 1)), (0, None)
    else:
        if j == 0:
            edge_neighbors = (0, (i-1, j, 2)), (3, None)
        else:
            edge_neighbors = (3, (i, j-1, 1)), (0, (i-1, j, 2))

    if not edge_neighbors[0]:
        return None

    edge_index, neighbor_indexes = edge_neighbors[0]
    i_n, j_n, e_n = neighbor_indexes
    edge = image[i_n][j_n].edges[e_n]
    rev_edge = reverse(edge)
    for tile_id, flip in edge_dict[rev_edge]:
        if tile_id in unavailable_tile_ids:
            continue
        tile = tile_dict[tile_id]
        if flip:
            tile = tile.flip()
        rotation = 4 + (tile.edges.index(rev_edge) - edge_index)
        tile = tile.rotate(rotation)
        edge_index, neighbor_indexes = edge_neighbors[1]
        edge = tile.edges[edge_index]
        rev_edge = reverse(edge)
        if not neighbor_indexes:
            if len(edge_dict[rev_edge]) == 1:
                return tile
        else:
            i_n, j_n, e_n = neighbor_indexes
            if image[i_n][j_n].edges[e_n] == rev_edge:
                return tile
    return None
    
def build_image(tiles, top_left, image_size):
    tile_dict = {tile.id: tile for tile in ALL_TILES}
    edge_dict = make_edge_dict(tiles)
    unavailable = {top_left}
    image = [[None] * image_size for i in range(image_size)]
    image[0][0] = tile_dict[top_left]
    try:
        for i in range(image_size):
            for j in range(image_size):
                tile = find_tiles(image, i, j, tile_dict, edge_dict, unavailable)
                if tile:
                    image[i][j] = tile
                    unavailable.add(tile.id)
                elif i != 0 or j != 0:
                    raise Exception(f"Cant find tile {i}, {j}!")
    except Exception:
        for row in image:
            print([t.id if t else 0 for t in row])
        global IMAGE
        IMAGE = image
        raise
    return image

def rotate_image(image, steps):
    its = [(0, 1), (1, 0), (0, -1), (-1, 0)]
    n = len(image)
    origins = [(0, 0), (0, n - 1), (n - 1, n - 1), (n - 1, 0)]
    it0 = its[(steps + 1) % 4]
    it1 = its[steps % 4]
    origin = origins[steps % 4]

    result = []
    for i in range(n):
        row = []
        result.append(row)
        pos = origin[0] + i * it0[0], origin[1] + i * it0[1]
        for j in range(n):
            row.append(image[pos[0]][pos[1]].rotate(steps))
            pos = pos[0] + it1[0], pos[1] + it1[1]
    return result

IMAGE = build_image(ALL_TILES, 2281, 12)
```

```python
def get_bitmap(image):
    img_size = len(image)
    tile_size = len(image[0][0].values)
    bits = [0] * img_size * tile_size
    for i in range(img_size):
        for j in range(img_size):
            values = image[i][j].values
            for k in range(tile_size):
                byte = int("".join("1" if c == "#" else "0" for c in values[k]), 2)
                bits[i * tile_size + k] = (bits[i * tile_size + k] << 8) + byte
    return bits

def find_monsters(bitmap):
    monster = ["00000000000000000010",
               "10000110000110000111",
               "01001001001001001000"]
    monster_mask = [int(s, 2) for s in monster]
    #roughness_mask = [int("".join("1" if c == "0" else "0" for c in s), 2) for s in monster]
    #result = []
    count = 0
    size = len(bitmap)
    for i in range(size - len(monster)):
        for j in range(size - 20):
            for k in range(len(monster)):
                if (bitmap[i+k] >> j) & monster_mask[k] != monster_mask[k]:
                    break
            else:
                count += 1
                #for k in range(len(monster)):
                #    result.append((bitmap[i+k] >> j) & roughness_mask[k])
    #return result
    return count
                    
BITS = get_bitmap(rotate_image(IMAGE, 3))
n_monsters = find_monsters(BITS)
n_bits = 0
for row in BITS:
    n_bits += bin(row)[2:].count("1")
print(n_bits - n_monsters * 15)
```

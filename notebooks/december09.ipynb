```python
def find_sum(n, values):
    values.sort()
    for i in range(len(values) - 1, -1, -1):
        if values[i] < n // 2:
            return False
        for j in range(i - 1, -1, -1):
            if values[i] + values[j] == n:
                return True
    return False
```

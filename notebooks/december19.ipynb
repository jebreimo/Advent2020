```python
def parse_rule_string(s):
    key, values = s.split(": ")
    if values[0] == '"':
        return int(key), "s", values[1]
    else:
        indexes = [[int(n) for n in q] for q in (p.split() for p in values.split(" | "))]
        return int(key), "i", indexes   
```

# Test part 1

```python
RULE_STRINGS = sorted("""0: 4 1 5
1: 2 3 | 3 2
2: 4 4 | 5 5
3: 4 5 | 5 4
4: "a"
5: "b\"""".split("\n"))
```

```python
MESSAGES = """ababbb
bababa
abbbab
aaabbb
aaaabbb""".split("\n")
```

# Part 1

```python
def make_validation_rules(rule_strings):
    rules = {}
    for s in rule_strings:
        index, kind, value = parse_rule_string(s)
        rules[index] = (kind, value)
    return rules          
RULES = make_validation_rules(RULE_STRINGS)
```

```python
def satisfies_rule(rules, rule, string):
    def _rec_check(rule_dict, in_rules, string):
        #print(in_rules, string)
        if not in_rules:
            return len(string) == 0
        elif not string:
            return False
        elif len(in_rules) > len(string):
            return False
        kind, next_rules = rule_dict[in_rules[-1]]
        if kind == "s":
            if string[0] != next_rules:
                return False
            return _rec_check(rule_dict, in_rules[:-1], string[1:])
        else:
            for next_rule in next_rules:
                out_rules = in_rules[:-1]
                out_rules.extend(reversed(next_rule))
                if _rec_check(rule_dict, out_rules, string):
                    return True
        return False
    return _rec_check(rules, [rule], string)
```

```python
sum(1 if satisfies_rule(RULES, 0, msg) else 0 for msg in MESSAGES)
```
